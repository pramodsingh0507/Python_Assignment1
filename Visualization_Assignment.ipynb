```python
# Ques 5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap


# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n = 100  # Number of samples
mean = [0, 0, 0]  # Mean of the features
cov = [[1, 0.8, 0.5],  # Covariance matrix
       [0.8, 1, 0.3],
       [0.5, 0.3, 1]]

# Generate samples from a multivariate normal distribution
data = np.random.multivariate_normal(mean, cov, n)

# Convert to a DataFrame
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

# Calculate the correlation matrix
corr = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
# Use another method

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100  # Number of samples
# Generate correlated features
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 0.5, n_samples)  # Correlated with feature_1
feature_3 = feature_1 * 0.5 + feature_2 * 0.5 + np.random.normal(0, 0.5, n_samples)  # Correlated with both

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Calculate the correlation matrix
corr = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()

```

```python
# Ques 4 Using the given x and y data, generate a 3D surface plot to visualize the function.


import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update the layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))

# Show the plot
fig.show()
```

```python
# Ques 5  Create a Bokeh heatmap using the provided dataset.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource

# Output to notebook (if you're using a Jupyter Notebook)
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh
data_source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    values=data_heatmap.flatten(),
))

# Create a color mapper
mapper = linear_cmap('values', palette='Viridis256', low=min(data_heatmap.flatten()), high=max(data_heatmap.flatten()))

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=400, height=400)

# Add squares to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=data_source, line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

```python
# Using Another method

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource

# Output to notebook (if you're using a Jupyter Notebook)
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.arange(10)  # Use indices for x
y = np.arange(10)  # Use indices for y

# Flatten the data for Bokeh
data_source = ColumnDataSource(data=dict(
    x=np.repeat(x, 10),  # Repeat x values
    y=np.tile(y, 10),    # Tile y values
    values=data_heatmap.flatten(),
))

# Create a color mapper
mapper = linear_cmap('values', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X Index', y_axis_label='Y Index', width=400, height=400)

# Add rectangles to the figure
p.rect(x='x', y='y', width=1, height=1, source=data_source, line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
